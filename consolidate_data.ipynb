```python
import pandas
pandas.set_option('display.max_columns', None)
```

```python
standings = pandas.read_csv("./data/all_teams_standings.csv")
standings.sample(3)
```

```python
awards = pandas.read_csv("./data/all_mvp_awards.csv")
awards_col = ["MVP_VOTES_SHARE", "MVP_WINNER", "MVP_PODIUM", "MVP_CANDIDATE"]
awards = awards[["player_season_team"] + awards_col]
awards.sample(3)
```

```python
stats = pandas.read_csv("./data/all_players_stats.csv")
stats.sample(3)
```

```python
all_data = stats.merge(awards, how='left', on="player_season_team")

for col in awards_col:
    all_data.loc[:, col] = all_data[col].fillna(False)
```

```python
all_data.sample(3)
```

```python
all_data = stats.merge(standings, how='inner', on=["TEAM", "SEASON"])
```
